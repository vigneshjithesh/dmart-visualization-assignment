```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
```

```python
df = pd.read_csv('DMart.csv')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
#find the null value 
df.isna().sum()
```

```python
# fill the null value with the help of fillna function and we got 0 null value 
df['Name'] = df['Name'].fillna('Premia Badam Almonds')
df['Brand'] = df['Brand'].fillna('Premia')
df['Price'] = df['Price'].fillna(109)
df['DiscountedPrice'] = df['DiscountedPrice'].fillna(85)
df['Category'] = df['Category'].fillna('Grocery')
df['SubCategory'] = df['SubCategory'].fillna('Grocery/Dry Fruits')
df['Quantity'] = df['Quantity'].fillna(200)
df['Description'] = df['Description'].fillna('USA')
df['BreadCrumbs'] = df['BreadCrumbs'].fillna('Grocery > Grocery/Dry Fruits')
df.isna().sum()
```

```python
# Distribution of Prices using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count plot of Categories
plt.figure(figsize=(10, 6))
sns.countplot(x='Category', data=df, palette='pastel')
plt.title('Count of Products in Each Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pairplot for numeric columns
sns.pairplot(df[['Price', 'DiscountedPrice']])
plt.show()
```

```python
# Scatter plot of Price vs DiscountedPrice
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price', y='DiscountedPrice', data=df, hue='Category', palette='Set2')
plt.title('Price vs DiscountedPrice')
plt.xlabel('Price')
plt.ylabel('DiscountedPrice')
plt.show()
```
